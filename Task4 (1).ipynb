```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("/content/customer_churn_dataset-testing-master.csv")

print("First 5 Rows:")
print(df.head())


print("\nDataset Info:")
print(df.info())

categorical_cols = df.select_dtypes(include=['object']).columns

print("\nCategorical Columns:")
print(categorical_cols)


label_encoder = LabelEncoder()


print("\nAfter Label Encoding:")


df_encoded = pd.get_dummies(df, drop_first=True)

print("\nDataset After One-Hot Encoding:")
print(df_encoded.head())

numerical_cols = ['Age', 'Tenure', 'Usage Frequency', 'Support Calls',
                  'Payment Delay', 'Total Spend', 'Last Interaction']

standard_scaler = StandardScaler()

df_standardized = df_encoded.copy()

df_standardized[numerical_cols] = standard_scaler.fit_transform(
    df_standardized[numerical_cols]
)

print("\nStandardized Data:")
print(df_standardized[numerical_cols].head())



minmax_scaler = MinMaxScaler()

df_normalized = df_encoded.copy()

df_normalized[numerical_cols] = minmax_scaler.fit_transform(
    df_normalized[numerical_cols]
)

print("\nNormalized Data:")
print(df_normalized[numerical_cols].head())


plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.hist(df['Age'], bins=20)
plt.title("Before Scaling")

plt.subplot(1,3,2)
plt.hist(df_standardized['Age'], bins=20)
plt.title("After Standardization")

plt.subplot(1,3,3)
plt.hist(df_normalized['Age'], bins=20)
plt.title("After Normalization")

plt.tight_layout()
plt.show()



df_standardized.to_csv("adult_standardized.csv", index=False)
df_normalized.to_csv("adult_normalized.csv", index=False)

print("\nFiles Saved Successfully!")
```
